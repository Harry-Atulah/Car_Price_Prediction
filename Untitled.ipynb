# Predicting Car Prices Using machine Learning

## Overview

This project aims to predict the prices of cars based on various features such as model, year, transmission type, mileage, fuel type, tax, mpg, and engine size. By leveraging machine learning techniques, we aim to develop a model that can provide accurate price predictions, thereby assisting buyers and sellers in making informed decisions. The project will involve data preprocessing, exploratory data analysis, model training, evaluation, and visualization to achieve this goal.

### Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

### Loading dataset

```python
df = pd.read_csv('ford.csv')
df.head()
```

```python
df.shape
```

```python
# Checking basic information about the dataset
df.info()
```

### Data Cleaning

```python
#Checking for missing values
df.isnull().sum()
```

The dataset has no missing values, which is excellent. This allows us to proceed without needing to handle missing data.

#### Standardizing

```python
# numerical features 
numerical_features = ['year', 'mileage', 'tax', 'mpg', 'engineSize']

# Initialize the scaler
scaler = StandardScaler()

# Fit the scaler on the numerical features and transform them
df[numerical_features] = scaler.fit_transform(df[numerical_features])
```

#### One Hot Encoding

```python
dataset_encoded = pd.get_dummies(df, columns=['model', 'transmission', 'fuelType'])
dataset_encoded.head()
```

### EDA

```python
#checking summary statistics
df.describe()
```

```python
# List of numerical features
numerical_features = ['year', 'mileage', 'tax', 'mpg', 'engineSize']

# histogram
dataset_encoded[numerical_features].hist(bins=15, figsize=(15, 6), layout=(2, 3))
plt.tight_layout()
plt.show()
```

```python
# Plot box plots
plt.figure(figsize=(15, 6))
for i, feature in enumerate(numerical_features, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=dataset_encoded[feature])
plt.tight_layout()
plt.show()
```

```python
# Scatter plots
plt.figure(figsize=(15, 10))
for i, feature in enumerate(numerical_features, 1):
    plt.subplot(2, 3, i)
    plt.scatter(dataset_encoded[feature], dataset_encoded['price'])
    plt.xlabel(feature)
    plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

```python
# box plots 
categorical_features = ['model', 'transmission', 'fuelType']

plt.figure(figsize=(15, 20))
for i, feature in enumerate(categorical_features, 1):
    plt.subplot(3, 1, i)
    sns.boxplot(x=df[feature], y=df['price'])
    plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
